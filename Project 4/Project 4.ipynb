```python
import urllib.request
rsp = urllib.request.urlopen('http://opendata.epa.gov.tw/ws/Data/REWXQA/?$orderby=SiteName&$skip=0&$top=1000&format=csv')
data=rsp.read()
csvdata = data.decode('utf8')
print(csvdata)
```

```python
air_dicts = {}
for line in csvdata.splitlines()[1:]:
    data = line.split(',')
    sitename, county, PSI, majorpollutant, status, publishtime = data[0], data[1], data[2], data[3], data[4], data[16]
    air_dicts[sitename] = [PSI, sitename, county, status, majorpollutant]
print(air_dicts)

print ('-----Sort by PSI-----')
for key, value in reversed(sorted(air_dicts.items(), key = lambda item:(item[1][0], item[0]))):
    print(key,':',value)
```
